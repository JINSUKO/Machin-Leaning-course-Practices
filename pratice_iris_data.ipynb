판다스로 변경

```python
import pandas as pd

X_iris = pd.DataFrame(iris.data);
y_iris = iris.target

X_iris_columns = iris['feature_names']
X_iris_types = ['setosa', 'versicolor', 'virginica']


X_iris[X_iris_columns[0]] = X_iris[0]
X_iris[X_iris_columns[1]] = X_iris[1]
X_iris[X_iris_columns[2]] = X_iris[2]
X_iris[X_iris_columns[3]] = X_iris[3]




X_iris_setosa = X_iris[y_iris == 0]
y_iris_setosa = y_iris[y_iris == 0]

X_iris_versicolor = X_iris[y_iris == 1]
y_iris_versicolor = y_iris[y_iris == 1]

X_iris_virginica = X_iris[y_iris == 2]
y_iris_virginica = y_iris[y_iris == 2]
```

```python
y_iris[y_iris == 0]
X_iris[y_iris == 0]
```

산점도 그래프 그려보기

```python
fig, axis = plt.subplots(2, 2)

cnt = 0
for i, row in enumerate(axis):
    for j, cell in enumerate(row):
        cell.set_xlabel('type')
        cell.set_ylabel(X_iris_columns[cnt])
        cell.scatter(x=y_iris[y_iris == 0], y=X_iris[y_iris == 0][cnt])
        cell.scatter(y_iris[y_iris == 1], X_iris[y_iris == 1][cnt])
        cell.scatter(y_iris[y_iris == 2], X_iris[y_iris == 2][cnt])
        fig.tight_layout()
        cnt += 1


plt.show()
```

Boxplot 그래프 그려보기

```python
# boxplot그리기
# https://velog.io/@youjung/matplotlib%EC%9C%BC%EB%A1%9C-Box-plot-%EA%B7%B8%EB%A6%AC%EA%B8%B0
# 눈금 표시하기
# https://wikidocs.net/92089

fig, axis = plt.subplots(2, 2, sharey=True)
# fig, axis = plt.subplots(2, 2)

cnt = 0
for i, row in enumerate(axis):
    for j, cell in enumerate(row):
        # cell.set_xlabel('type')
        cell.set_ylabel(X_iris_columns[cnt])
        cell.boxplot([X_iris[y_iris == 0][cnt], X_iris[y_iris == 1][cnt], X_iris[y_iris == 2][cnt]])
        cell.set_xticks([1, 2, 3], [X_iris_types[0], X_iris_types[1], X_iris_types[2]])
        fig.tight_layout()
        cnt += 1


plt.show()
```

머신러닝 모델을 학습시키기 위해서 Feature Scaling 수행

표준화 진행

```python
# 표준화
from sklearn.preprocessing import StandardScaler

stdscaler = StandardScaler()

X_iris_str_column = X_iris[[X_iris_columns[0], X_iris_columns[1], X_iris_columns[2], X_iris_columns[3]]]

stdscaler.fit(X_iris_str_column)

X_scaled = stdscaler.transform(X_iris_str_column)
```

train-test data 분리

```python
from sklearn.model_selection import train_test_split

X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y_iris, random_state=42)
```

정규화 진행

```python
from sklearn.preprocessing import MinMaxScaler

minmaxscaler = MinMaxScaler()

minmaxscaler.fit(X_iris_str_column)

X_minmax_scaled = minmaxscaler.transform(X_iris_str_column)

X_minmax_scaled
```

```python
X_train_minmax_scaled, X_test_minmax_scaled, y_train, y_test = train_test_split(X_minmax_scaled, y_iris, random_state=42)
```

```python
from sklearn.preprocessing import Normalizer

normalscaler = Normalizer()

normalscaler.fit(X_iris_str_column)

X_normal_scaled = normalscaler.transform(X_iris_str_column)

X_normal_scaled
```

```python
X_train_normal_scaled, X_test_normal_scaled, y_train, y_test = train_test_split(X_normal_scaled, y_iris, random_state=42)
```

KNN 분류기 수행

```python
from sklearn.neighbors import KNeighborsClassifier

knmodel = KNeighborsClassifier()

knmodel.fit(X_train_scaled, y_train)
X_train_scaled.shape
```

KNN 분류기 시험

```python
print(knmodel.score(X_train_scaled, y_train))
print(knmodel.score(X_test_scaled, y_test))

X_train_scaled[0]

knmodel.predict([X_test_scaled[0]])
```

```python
knmodel_minmax = KNeighborsClassifier()

knmodel_minmax.fit(X_train_minmax_scaled, y_train)
X_train_minmax_scaled.shape
```

```python
print(knmodel_minmax.score(X_train_minmax_scaled, y_train))
print(knmodel_minmax.score(X_test_minmax_scaled, y_test))

knmodel_minmax.predict([X_test_minmax_scaled[0]])
```

```python
knmodel_normal = KNeighborsClassifier()

knmodel_normal.fit(X_train_normal_scaled, y_train)
X_train_normal_scaled.shape
```

```python
print(knmodel_normal.score(X_train_normal_scaled, y_train))
print(knmodel_normal.score(X_test_normal_scaled, y_test))

knmodel_normal.predict([X_test_normal_scaled[0]])
```
