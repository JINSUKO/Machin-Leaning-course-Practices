Iris 데이터

```python
from sklearn.datasets import load_iris
iris = load_iris()

# https://blog.naver.com/flunatic/222155349609 그래프 간격 조절 설명 사이트
```

```python
import matplotlib.pyplot as plt
```

판다스로 변경

```python
import pandas as pd

X_iris = pd.DataFrame(iris.data);
y_iris = iris.target

X_iris[X_iris_columns[0]] = X_iris[0]
X_iris[X_iris_columns[1]] = X_iris[1]
X_iris[X_iris_columns[2]] = X_iris[2]
X_iris[X_iris_columns[3]] = X_iris[3]


X_iris_columns = iris['feature_names']
X_iris_types = ['setosa', 'versicolor', 'virginica']


X_iris_setosa = X_iris[y_iris == 0]
y_iris_setosa = y_iris[y_iris == 0]

X_iris_versicolor = X_iris[y_iris == 1]
y_iris_versicolor = y_iris[y_iris == 1]

X_iris_virginica = X_iris[y_iris == 2]
y_iris_virginica = y_iris[y_iris == 2]
```

```python
y_iris[y_iris == 0]
X_iris[y_iris == 0]
```

산점도 그래프 그려보기

```python
fig, axis = plt.subplots(2, 2)

cnt = 0
for i, row in enumerate(axis):
    for j, cell in enumerate(row):
        cell.set_xlabel('type')
        cell.set_ylabel(X_iris_columns[cnt])
        cell.scatter(x=y_iris[y_iris == 0], y=X_iris[y_iris == 0][cnt])
        cell.scatter(y_iris[y_iris == 1], X_iris[y_iris == 1][cnt])
        cell.scatter(y_iris[y_iris == 2], X_iris[y_iris == 2][cnt])
        fig.tight_layout()
        cnt += 1


plt.show()
```

Boxplot 그래프 그려보기

```python
# boxplot그리기
# https://velog.io/@youjung/matplotlib%EC%9C%BC%EB%A1%9C-Box-plot-%EA%B7%B8%EB%A6%AC%EA%B8%B0
# 눈금 표시하기
# https://wikidocs.net/92089

fig, axis = plt.subplots(2, 2, sharey=True)
# fig, axis = plt.subplots(2, 2)

cnt = 0
for i, row in enumerate(axis):
    for j, cell in enumerate(row):
        # cell.set_xlabel('type')
        cell.set_ylabel(X_iris_columns[cnt])
        cell.boxplot([X_iris[y_iris == 0][cnt], X_iris[y_iris == 1][cnt], X_iris[y_iris == 2][cnt]])
        cell.set_xticks([1, 2, 3], [X_iris_types[0], X_iris_types[1], X_iris_types[2]])
        fig.tight_layout()
        cnt += 1


plt.show()
```

```python
X_iris[[0,1,2,3]], y_iris
```

머신러닝 모델을 학습시키기 위해서 Feature Scaling 수행

```python
# 정규화
from sklearn.preprocessing import StandardScaler

stdscaler = StandardScaler()

X_iris_str_column = X_iris[[X_iris_columns[0], X_iris_columns[1], X_iris_columns[2], X_iris_columns[3]]]

stdscaler.fit(X_iris_str_column)

X_scaled = stdscaler.transform(X_iris_str_column)
```

train-test data 분리

```python
from sklearn.model_selection import train_test_split

X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y_iris, random_state=42)
```

KNN 분류기 수행

```python
from sklearn.neighbors import KNeighborsClassifier

knmodel = KNeighborsClassifier()

knmodel.fit(X_train_scaled, y_train)
X_train_scaled.shape
```

KNN 분류기 시험

```python
print(knmodel.score(X_train_scaled, y_train))
print(knmodel.score(X_test_scaled, y_test))

X_train_scaled[0]

knmodel.predict([X_test_scaled[0]])
```

PCA (주성분 분석, Principal Component Analysis)

원본 데이터를 차원 축소한 것과 정규화를 진행해서 나눈 데이터를 차원 축소한 것, 정규화를 진행 후 차원축소하여 나눈 데이터끼리의
2차원을 기준으로 산점도를 기준으로 시각화 해보았다.

원본데이터와 양상이 제일 비슷한 데이터는

 훈련 데이터와 테스트 데이터로 나누기 전에 차원을 축소한 데이터였다.
 
정규화한 데이터를 훈련데이터와 테스트데이터로 나눈 후 차원 축소한 그래프는 원본을 차원 축소한 데이터와 양상이 다르게 바뀌었기 때문에 사용할 수 없는 데이터로 생각할 수 있을 것같다.

훈련데이터와 테스트데이터로 나누기 전에 차원축소를 진행한 데이터의 그래프가 원본을 차원 축소한 데이터와 양상이 비슷하기 때문에 원본의 특성을 최대한 담은 신뢰할 수 있는 데이터라고 생각해 볼 수 있다.

```python
# 위의 두번째 그래프로 그려본 데이터를 KNN 분류기로 시험해봄.

knmodelpca = KNeighborsClassifier()

knmodelpca.fit(X_train_scaled_pca_d2, y_train)

print((knmodelpca.score(X_train_scaled_pca_d2, y_train)))
print((knmodelpca.score(X_test_scaled_pca_d2, y_test)))
```

```python
# 위의 두번째 그래프로 그려본 데이터를 KNN 분류기로 시험해봄.

knmodelpca = KNeighborsClassifier()

knmodelpca.fit(X_train_scaled_pca_d2_v2, y_train_v2)

print(X_test_scaled_pca_d2_v2[0])
print(X_test_scaled_pca_d2_v2[1])
print(X_test_scaled_pca_d2_v2[2])
print(knmodelpca.predict([[0.6328582, -0.41638772]]))
print(knmodelpca.predict([[-1.898572, 1.40501879]]))
print(knmodelpca.predict([[3.31069558, 0.01778095]]))

print((knmodelpca.score(X_train_scaled_pca_d2_v2, y_train_v2)))
print((knmodelpca.score(X_test_scaled_pca_d2_v2, y_test_v2)))
```
