```python
import pandas as pd

df = pd.read_csv('./penguins.csv')
df = df.drop(columns=['year'])
df
```

```python
df_bool1 = df['bill_length_mm'].isna()

for i, v in enumerate(df_bool1):
    if v == False:
        df_bool1[i] = True
    else:
        df_bool1[i] = False
        
df = df[df_bool1]

df.reset_index().drop(axis=1, columns=['index'])
```

```python
df_no_sex_data = df[df['sex'].isna()]

df_no_sex_data = df_no_sex_data.reset_index().drop(axis=1, columns=['index'])
df_no_sex_data
```

```python
df = df.dropna(how='any')
df = df.reset_index().drop(axis=1, columns=['index'])
df
```

rom sklearn.preprocessing import OneHotEncoder 시도하였는데,
변경된 데이터에서 입력한 데이터프레임과 순서가 맞지 않아서 쓰지 get_dummies 이용함.

--> DataFrame 변경 시 순서 문제가 발생한 것이었고, index=df_drop_island.index 파라미터를 사용하면 순서가 유지된다.

```python
df_drop_island = df.drop(columns=['island'])
df_drop_island

species_list = df_drop_island['species'].unique()

from sklearn.preprocessing import OrdinalEncoder

ordinalencoder = OrdinalEncoder(categories=[species_list])

print(species_list)

df_species = pd.DataFrame(ordinalencoder.fit_transform(df_drop_island[['species']]), index=df_drop_island.index)

df_drop_island['species'] = df_species.rename(columns={0:"species"})
df_drop_island
```

```python
df_drop_species = df.drop(columns=['species'])

island_list = df_drop_species['island'].unique()
print(island_list)

ordinalencoder = OrdinalEncoder(categories=[island_list])

df_island = pd.DataFrame(ordinalencoder.fit_transform(df_drop_species[['island']]), index=df_drop_species.index)

df_drop_species['island'] = df_island.rename(columns={0:"island"})
df_drop_species
```

성별 컬럼 one-hot encoding

```python
y_species = df_drop_island['species']
df_species_data = df_drop_island.drop(axis=1, columns=['species'])
df_species_data = pd.get_dummies(df_species_data, dtype='int64')
df_species_data


y_island = df_drop_species['island']
df_island_data = df_drop_species.drop(axis=1, columns=['island'])
df_island_data = pd.get_dummies(df_island_data, dtype='int64')
df_island_data
```

```python
y_sex = df_island_data['sex_female']
y_sex.rename({'sex_female':'sex'})

df_sex_data = df_island_data.drop(axis=1, columns=['sex_female', 'sex_male'])
```

```python
from sklearn.model_selection import train_test_split

X_species_train, X_species_test, y_species_train, y_species_test = train_test_split(df_species_data, y_species, random_state=42)
X_island_train, X_island_test, y_island_train, y_island_test = train_test_split(df_island_data, y_island, random_state=42)
X_sex_train, X_sex_test, y_sex_train, y_sex_test = train_test_split(df_sex_data, y_sex, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

for i in [1, 2, 3, 4, 5, 10, 15, 20, 50, 100]:
    dtClassifier = DecisionTreeClassifier(max_depth=i, random_state=42)
    
    dtClassifier.fit(X_species_train, y_species_train)
    
    print(f'train max_depth={i}:', dtClassifier.score(X_species_train, y_species_train))
    print(f'test max_depth={i}:', dtClassifier.score(X_species_test, y_species_test))
    print(dtClassifier.feature_importances_)
    print()
```

```python
for i in [1, 2, 3, 4, 5, 10, 15, 20, 50, 100]:
    dtClassifier = DecisionTreeClassifier(max_depth=i, random_state=42)
    dtClassifier.fit(X_island_train, y_island_train)
    
    print(f'train max_depth={i}:', dtClassifier.score(X_island_train, y_island_train))
    print(f'test max_depth={i}:', dtClassifier.score(X_island_test, y_island_test))
    print(dtClassifier.feature_importances_)
    print()
```

```python
for i in [1, 2, 3, 4, 5, 10, 15, 20, 50, 100]:
    dtClassifier = DecisionTreeClassifier(max_depth=i, random_state=42)
    dtClassifier.fit(X_sex_train, y_sex_train)

    print(f'train max_depth={i}:', dtClassifier.score(X_sex_train, y_sex_train))
    print(f'test max_depth={i}:', dtClassifier.score(X_sex_test, y_sex_test))
    print(dtClassifier.feature_importances_)
    print()
```
